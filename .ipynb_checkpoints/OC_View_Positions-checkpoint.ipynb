```python
import mrigutilities as mu
import pandas as pd
import json

def find_atm(a,diff):
    # diff = 2.5
    divisor = diff
    cnt = 1
    while divisor%5 != 0:
        divisor *=  cnt
        cnt += 1

    a1 = int(round(a/divisor)*divisor/divisor)*divisor
    b = a1
    if abs(a-b-diff) < diff/2:
        b = b + diff
    return b
```

```python
session = mu.getKiteSession()
niftyatm= session.quote('NSE:NIFTY 50')['NSE:NIFTY 50']['last_price']
print(niftyatm)
```

```python
pr= session.quote('NSE:SBIN')['NSE:SBIN']['last_price']
find_atm(pr,5)
```

```python
colorder = ['scrip','Long|orig_liab','Long|curr_liab','Long|T/D','Long|theta (T)','Long|delta (D)','Long|pnl','Long|qty','Long|instrument','strike','Short|instrument','Short|qty','Short|pnl','Short|delta (D)','Short|theta (T)','Short|T/D','Short|curr_liab','Short|orig_liab']
df1 = df[1][df[1]['l_s'] != ''].pivot(columns=['l_s']).swaplevel(axis=1)
df1.columns = df1.columns.map('|'.join).str.strip('|')
df1['scrip'] = df1['Long|scrip'].fillna('')  + df1['Short|scrip'].fillna('') 
df1['strike'] = df1['Long|strike'].fillna('') + df1['Short|strike'].fillna('')
df1.drop(columns=['Long|strike','Short|strike'],inplace=True)
df1.drop(columns=['Long|scrip','Short|scrip'],inplace=True)
print(df1)
df1.fillna(0,inplace=True)

df1_agg = df1.groupby(by=['scrip'], as_index=False)['Long|orig_liab','Long|curr_liab','Long|T/D','Long|theta (T)','Long|delta (D)','Long|pnl','Long|qty','Short|qty','Short|pnl','Short|delta (D)','Short|theta (T)','Short|T/D','Short|curr_liab','Short|orig_liab'].sum().fillna(0).round()#.assign([,'Long|instrument','strike','Short|instrument']='')


df1_agg.sort_index(axis=1)
df1_agg
df_new = (pd.concat([df1, df1_agg])
      .reindex(df1.columns, axis=1)
      .fillna('')
      .sort_values(['scrip','strike','Long|instrument','Short|instrument'], ascending=True, ignore_index=True))
df_new[colorder]
```
